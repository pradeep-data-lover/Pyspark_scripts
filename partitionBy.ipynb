```
df=spark.read.option('inferschema',True).option('header',True).option('sep',",").format('csv').load('/FileStore/tables/Popular_Baby_Names.csv')
display(df)
```

```
df.groupBy('Year of Birth').count().show(truncate=False)
```

```
df.groupBy('Year of Birth','Gender').count().show(truncate=False)
```

```
df.write.option('header',True)\
    .partitionBy('Year of Birth')\
        .mode('overwrite')\
            .csv('/FileStore/tables/output_partition_files')
```

```
df.write.option('header',True)\
    .partitionBy('Year of Birth','Gender')\
        .mode('overwrite')\
            .csv('/FileStore/tables/output_by_2col_partition_files')
```

```
df.write.option('header',True)\
    .partitionBy('Year of Birth')\
        .option('maxRecordsPerFile',4200)\
            .mode('overwrite')\
                .csv('/FileStore/tables/output_partition_files_withlimit')
```
